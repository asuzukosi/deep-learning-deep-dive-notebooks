```python
# Activation functions

# We use different activation functions for different cases.
# Understanding how different activation functions work can help you pick the right ones to
# use in specific situations.

# The activation function is applied to the output of a neuron or the set of outputs of a layer hehe.
# The activation function modifies the output. We use activation functions to apply non linarity to our neural network.

# It allows neural networks with two or more layers to map non-linear fucntions. 

# There are generally 2 types of activation functions, those used on hidden layers,
# and those used on output layers.

# Usually, the activation function used in the hidden layer will be the same for all of them
# but they dont have to
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The step activation function
# The purpose of the step activation function is to mimic neuron firing and not firing. 
# based on input information. 

# The simplest version of the step activation function is to fire if the output is 1 and dont fire
# if its less than 1

# The step function was previously frequently used in hidden layers, but now
# it barely ever a choice.
def step_activation(x):
    out = []
    for i in x:
        if i > 0:
            out.append(1)
        else:
            out.append(0)
            
    return np.array(out)
        

x = np.arange(-8, 8, 0.1)
y = step_activation(x)

plt.plot(x,y)
plt.title("Step activation")
plt.show()
```

```python
# The linear actiavtion function
# The linear actiavtion function is simply the equeation of a line. 
# It will appear as a linear straight line with y=x
# The linear actiavtion function is usually applied to the last layer of a network in the case
# of a linear network, but its effects are not substatial
```

```python
# The Sigmoid activation function
# The problem with the step function is that it is not very informative. 
# Because it is either 1 (good) or 0 (bad), we can not determin 'how' good or bad
# the network is.
# The sigmoid activation function is more granular and informative

def simoid(x):
    return 1/(1 + np.exp(-x))

y = simoid(x)

plt.plot(x, y)
plt.title("Sigmoid activation function")
plt.show()
```

```python
# The sigmoid fucntion returns a value between 0 and 1. The more positive a value is
# The closer it tends to 1, while the more negative a value is the closer it tends to zero.

# 'dead nuerons' are neurons that can not be trained because either their values or 
# derivatives are stuck at 0.

# In the sigmoid activation the result contains all the properties of its input. 
# The importance of non-linearity to be explained further soon.

# The sigmoid activation was eventually replaced for use in the hidden layers by the 
# ReLU activation function, because the ReLU activation function was better for training
# because it was less likely to experience the vanishing gradient problem. 
```

```python
# Rectified linear unit activation function
# The rectified linear unit is similar to the linear activation function but sets all 
# values that are less than zero to zero. 

def rectifiedLinearUnitActivation(x):
    out = []
    for i in x:
        if i < 0:
            out.append(0)
        else:
            out.append(i)
            
    return out

y = rectifiedLinearUnitActivation(x)
y
plt.plot(x, y)
plt.title("Rectified linear unit activation")
plt.show()
```

```python
# The rectified linear unit is simpler than the sigmoid function, but its more complex than the linear function
# If x is less than 0 then y is 0, else y is x
# This simple yet powerful activatioinfunction is the most used activation function at the time of reading this book was written
# The sigmoid function isn't the most complicated but it is harder to train that the ReLU. 
# The ReLU is very close to being  a linear activation function but still remains non linear. This simple property is however very effective
```

```python
# We would be implementing the ReLU activation function as a layer
class Activation_ReLU:
    def __init__(self):
        pass
    
    def forward(self, X):
        # We would calculate the ReLU activation of a value
        # Checking the maximum between each value and 0
        self.output = np.maximum(0, X)
        return self.output
```

```python
# Implementing the dense layer again
class Layer_Dense:
    def __init__(self, n_inputs, n_outputs):
        # We will multiply the results by 0.01 to scale the values down fo easier training. 
        self.weights = 0.01 * np.random.randn(n_inputs, n_outputs) # We shape it this way so it would be easy for matrix
                                                            # multiplication as we move through the layers, i.e the num_inputs,
                                                            # # would match with the first 
                                                            # matrix which would have the inputs arranged as row vectors of length m
        self.biases = np.zeros(shape=(1, n_outputs)) # we initialize our biases to zero to prevent the possibility of dead neurons
                                                      
    def forward(self, X):
        self.output = np.dot(X, self.weights) + self.biases
        return self.output
```

```python
# The sigmod activation function
# In our case we are looking for our model to be a classifer. So we would 
# like to use an activation function for classification. one of these is the softmax activation function.

# Why do we need another activation function. This depends on our overall goal
# In regular activation functions, the result is exclusive meaning that the each output
# from the output function is independent

# To address this lack of context, the softmax activation function can take in a non-calibrated set of inputs and
# produce a normal distribution from of probabilities from it. In the case of classification, 
# We want it to be a probability of certainty for each class. The values returned by the softmax function reveals the 
# confidence score of our network for a specific class. To think deeply about it, we dont really need to know the 
# actual values of each class, we just want to see the probability distribution across other classes. 
# The softmax acivation fucntion then returns the confidence scores for each of hte clases

# Therefore, the predicted class is the class with the largest confidence score.
```

```python
# The softmax function
layer_outputs = [[4.8, 1.21, 2.385]]
# First we need to exponentiate the output using Euler's number e, which is roughly
# 2.718228 this number is known as exponential growth number. 
# y = exp(x)
exponentiated_values = np.exp(np.array(layer_outputs))
exponentiated_values
```

```python
# The exponential serves two purposes: 
# 1. if we want to solve for probabilities then we need non negative values.

# Ultimately we want to convert the outputs of the layer into a probability distribution
# that all sums up to one. 
sum_exponentiated = np.sum(exponentiated_values, axis=1, keepdims=True)
sum_exponentiated
```

```python
softmax_distribution = exponentiated_values / sum_exponentiated
softmax_distribution
```

```python
# Implementationf of the softmax activation function layer. 
class Activation_Softmax:
    def __init__(self):
        pass
    
    def forward(self, X):
        exp_values = np.exp(X - np.max(X, axis=1, keepdims=True)) # We subtracted the values of the outputs by the largest value to scale down the values.
                                                                  # large numbers and dead neurons are problems that perturb neural networks.
                                                                  # The exponential function used in the softmax activation is one of the causes of exploding values
                                                                  # While the sigmoid activaiton i sone of the causes of vanishing gradient.
                                                                  # Using exponent of extremely large values can cause an overflow error, so we have to subtract them by the largest values
                                                                  
        probabilities = exp_values / np.sum(exp_values, axis=1, keepdims=True)
        self.output = probabilities
        return self.output
```
