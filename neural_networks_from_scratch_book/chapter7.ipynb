```python
# Derivatives

# Randomly changing and searching for optimal weigths and biases will not provide a
# solution for finding the best set of parameters to reduce the loss function. 

# The number of combinations are just to large and it would be inefficient to explore
# all the various combinations of parameters. 

# Each weight and bias must have varying degress of influence to the loss. This influece
# is determined by the value of parameters and the input values. 

# The function of how the parameters affect the loss is not neccessary linear. To know how 
# to adjust the parameters, we need to understand their impact on the loss. 

# Althogh the weights and biases are not direct parameters to the loss function, their values
# affect the output of the network which is in turn passed to the loss function. 
```

```python
# To undertand partial derivatives, we must first understand derivatives
# Derivatives are calculatef from a function with a single parameter
```

```python
# Impact of a parameter on the output
# To understand what we mean by impact, let's start with looking at a single function
def f(x):
    return 2*x
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(5)
y = f(x)
```

```python
plt.plot(x, y)
plt.title("Sample display")
plt.show()
```

```python
# How can we define the impact of x on y? 
# It can be said that y is double of x
# Slope can be defined as the change in y/change in x. In math --- delta y / delta x

# To calculate the slope, we will take the values at any two points in the slope and calculate 
# them

pointA  = np.array([0, 0])
pointB = np.array([1, 2])

result = pointB - pointA
result[1] / result[0]
```

```python
# creating a polynomial function
def f(x):
    return 2*x**2
```

```python
y = f(x)

plt.plot(x, y)
plt.show()
```

```python
# Since this grapth is not linear, the slope changes at different points
# ie the slope is not the same throughout the graph
# This is also known as a parabolic function


# The question now becomes how can we measure the impact x has on y in a non linear function
# Calculus provides a method called the instantaneous slope. (Slope at a specific point), which is known as 
# the derivative

# The tangent line is made by drawing a line that is extremely small across two points in a slope
# There is no 'single' slope, the slope depends on where you measure it.

# To experiment with this concept, we will measure two values of x that are a very small distance apart. i.e 1e-4
```

```python
x = 1
y1 = f(x)
y2 = f(x + (1e-4))
(y2-y1)/1e-4
```

```python
# Trying it at a different point will produce a different result.
x = 2
y1 = f(x)
y2 = f(x + (1e-4))
(y2-y1)/1e-4
```

```python
# This is called finding the derivative with the tangent. 
# We choose a rounding error small enough to give us an approximate of the derivative
# and large enough not to be regarded as a rounding error.
```

```python
# The numerical derivative
# The method we just used to calculate the derivative is called the numerical derivative.
# It uses the slope and tangent line and finds two points that are infinitely close and calculates the
# slope between them. 

# Let's take a more granular look at the function to better understand the numerical derivative
x = np.arange(0, 5, 0.001)
y = f(x)

plt.plot(x, y)
plt.show()
```

```python
approximage_derivative = (y2 - y1)/ 1e-4
b = y2 - approximage_derivative*(2+1e-4)
b
```

```python
def tangent_line(x):
    return approximage_derivative*x + b


to_plot = [2-0.9, 2, 2+0.9]
plt.plot(to_plot, [tangent_line(i) for i in to_plot])
x = np.arange(0, 5, 0.001)
y = f(x)

plt.plot(x, y)
plt.show()
```

```python
# This tells us the instantanous rate of change of a particular value of x on our output y
```

```python
# We will use similar concepts to identify the rate of change in our loss based on our parameters. 
```

```python
# Lets experiment the same code on various points in our line
x = np.arange(0,5, 0.001)
y = f(x)
plt.plot(x, y)
colors = ['k', 'g', 'r', 'b', 'c']

def approximate_tangent_line(x, approximage_derivative):
    return (approximage_derivative * x) + b

for i in range(5):
    p2_delta = 1e-4
    x1 = i
    x2 = x1 + p2_delta
    y1 = f(x1)
    y2 = f(x2)
    
    print((x1, y1), (x2, y2))
    approximage_derivative = (y2 - y1)/(x2 - x1)
    b = y2 - (approximage_derivative*x2)
    
    to_plot = [x1-0.9, x1, x1+0.9]
    plt.scatter(x1, y1, c=colors[i])
    plt.plot([point for point in to_plot], [approximate_tangent_line(point, approximage_derivative) for point in to_plot], c=colors[i])
    print("Approximate derivative for f(x)", f"where x = {x1} is {approximage_derivative}")

plt.show()
```
