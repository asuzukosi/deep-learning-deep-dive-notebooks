```python
import sklearn
from sklearn.datasets import load_digits
```

```python
(x, y) = load_digits(return_X_y=True)
x_train, x_test = x[:1700], x[1700:]
y_train, y_test = y[:1700], y[1700:]
```

```python
x_test.shape
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
example = x[0]
np.sum(np.abs(x - example), axis=1).shape
```

```python
class NearestNeighborImageClassifier:
    
    def train(self, X, y):
        self.Xtr = X
        self.ytr = y
        
    def predict(self, X):
        distances = np.sum(np.abs(self.Xtr - X), axis=1)
        index = np.argmin(distances)
        ypred = self.ytr[index]
        return ypred
```

```python
classifier = NearestNeighborImageClassifier()
classifier.train(x_train, y_train)
```

```python
for index in range(97):
    print("The actual value is  ", y_test[index])
    print("The predicted value is ", classifier.predict(x_test[index]))
```

```python
class KNearestNeighborsImageClassifier:
    def __init__(self, n_neighbors) -> None:
        self.neighbors = n_neighbors
        
    def train(self, x, y):
        self.Xtr = x
        self.Ytr = y
    
    def predict(self, X):
        distances = np.sum(np.abs(self.Xtr - X), axis= 1)
        predictions = dict()
        for i in range(self.neighbors):
            index= np.argmin(distances)
            value  = self.Ytr[index]
            if value not in predictions:
                predictions[value] = 1
            else:
                predictions[value] += 1
        return predictions
```

```python
classifier = KNearestNeighborsImageClassifier(5)
classifier.train(x_train, y_train)
```

```python
for index in range(97):
    print("The actual value is  ", y_test[index])
    print("The predicted value is ", classifier.predict(x_test[index]))
```
